```python
import pandas as pd
import requests
import plotly.express as px


# ================================
# Somar exames por município
# ================================

mapa_mamografia = (
    df_geral
    .groupby('municipio', as_index=False)
    ['valor_indicador']
    .sum()
)


# ================================
# Baixar GeoJSON dos municípios CE
# ================================

url = "https://raw.githubusercontent.com/tbrugz/geodata-br/master/geojson/geojs-23-mun.json"

geo_ce = requests.get(url).json()


# ================================
# Criar mapa
# ================================

fig = px.choropleth_mapbox(
    mapa_mamografia,
    geojson=geo_ce,
    locations="municipio",
    featureidkey="properties.name",
    color="valor_indicador",
    color_continuous_scale="YlOrRd",
    hover_name="municipio",
    hover_data={
        "valor_indicador": ":,"
    },
    mapbox_style="carto-positron",
    center={
        "lat": -5.2,
        "lon": -39.5
    },
    zoom=6,
    opacity=0.75
)


fig.update_layout(
    title="Distribuição de mamografias realizadas por município - Ceará",
    height=800
)


fig.show(renderer="browser")
```
