# Caso de dados inconsistentes

- Valores em 0.0 não tiveram envio de dados para o ano
- Em 2022 houve consolidação maior do uso do e-SUS APS 
- Submissão tardia de dados
- Perda de dados durante a migração
- Acumulo e dados para demais anos (Farias Brito 202212 - Compensando demais anos anteriores).
- Utilização de dois sistemas paralelos para registro de informações (SIAB e e-SUS APS) 
- SIAB passou a ser abolido em 2017 de forma obrigatoria 
- e-SUS APS implantado em 2013 e obigatório em 2017
- Em muitos municípios, o exame é coletado na UBS (APS) mas registrado no SISCAN
- Dados de cancer de colo de utero podem ter registrado em demais sitemas além do e-SUS APS, como o SISCAN.
- e-SUS registra apenas parte dos exames (profissionais), a partir disso o restante vai para SISCAN
- SISCAN e e-SUS são complementares e não se sobrepoem

# Mamografia de 50 a 69 anos


```python
df_mama = pd.read_csv("../../Datasets/N1/ccmec50a69.csv")

df_mama
```

```python
df_cariri_mama = df_mama[
    (df_mama['sg_uf'] == 'CE') &
    (df_mama['no_municipio'].isin(cidades)) &
    (df_mama['vl_indicador_calculado_mun'] != 0)
]

df_filtrado_mama = df_cariri_mama.groupby('no_municipio').size().reset_index(name='quantidade')

df_filtrado_mama

df_cariri_mama
```

# Geração de datasets filtrados.

Os seguintes datasets tiveram: 
- Removidas linhas que continham valores 0.0 na coluna vl_indicador_calculado_mun.
- Mantidas apenas linhas que fazem parte da região do Cariri, no ceará.

As operações ocorreram nos datasets ccmec50a69.csv e ccmec25a64_main.csv, que correspondem respectivamente a dados de exames citopatológicos e exames de mamografia.

Seguintes colunas foram mantidas (origem):
- sg_uf
- no_municipio
- vl_indicador_calculado_mun
- co_anomes
- dt_atualizacao

Colunas renomeadas para: 
- estado
- municipio
- valor_indicador
- ano_mes
- data_atalizacao

As seguintes colunas foram adicionadas:

- tipo_exame: informa o tipo de exame realizado (mamografia/citopatologia)


```python
colunas = ['sg_uf', 'no_municipio', 'vl_indicador_calculado_mun', 'co_anomes', 'dt_atualizacao']

mapa_nomes = {
    'sg_uf': 'estado',
    'no_municipio': 'municipio',
    'vl_indicador_calculado_mun': 'valor_indicador',
    'co_anomes': 'ano_mes',
    'dt_atualizacao': 'data_atualizacao'
}

df_citopatologia_cariri = df_filtrado[colunas].rename(columns=mapa_nomes)
df_mamografia_cariri = df_cariri_mama[colunas].rename(columns=mapa_nomes)

df_citopatologia_cariri['tipo_exame'] = 'Citopatologico'
df_mamografia_cariri['tipo_exame'] = 'Mamografia'

df_citopatologia_cariri.to_csv("../../Datasets/N1/citopatologia_cariri.csv", index=False)
df_mamografia_cariri.to_csv("../../Datasets/N1/mamografia_cariri.csv", index=False)
```

### Criado novo dataset

Juntando os dataframes df_citopatologia_cariri e df_mamografia_cariri, foi criado um novo dataset chamado cito_e_mamo_cariri. <br>
Mantendo apenas os dados da região do cariri, descrevendo o tipo de exame realizado.

```python
df_cito_e_mamo = pd.concat(
    [df_citopatologia_cariri, df_mamografia_cariri],
    ignore_index=True
)

df_cito_e_mamo = df_cito_e_mamo[df_cito_e_mamo['ano_mes'] != 202601]

df_cito_e_mamo.to_csv('../../Datasets/N1/cito_e_mamo_cariri.csv', index=False)
```
