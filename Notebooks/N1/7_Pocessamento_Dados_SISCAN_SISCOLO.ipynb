# Processamento de dados para novas análises
---
Novo processo sob datasets mamografia_atendbrCE e cito_colo_atendbrCE

Será gerado um novo dataset, que será o tratamento e a fusão dos dados em um só

```python
# Imports

import pandas as pd
import seaborn as sns
```

```python
# Carregando dados

# Cidades do cariri
cidades_cariri = [
    "ABAIARA", "ALTANEIRA", "ANTONINA DO NORTE", "ARARIPE", "ASSARE",
    "AURORA", "BARBALHA", "BARRO", "BREJO SANTO", "CAMPOS SALES",
    "CARIRIACU", "CRATO", "FARIAS BRITO", "GRANJEIRO", "JARDIM",
    "JATI", "JUAZEIRO DO NORTE", "MAURITI", "MILAGRES", "MISSAO VELHA",
    "NOVA OLINDA", "PENAFORTE", "PORTEIRAS", "POTENGI", "SALITRE",
    "SANTANA DO CARIRI", "TARRAFAS", "VARZEA ALEGRE"
]

# Datasets
df_cito_siscan = pd.read_csv("../../Datasets/N1/cito_colo_atendbrCE.csv", delimiter=";") 
df_mamo_siscan = pd.read_csv("../../Datasets/N1/mamografia_atendbrCE.csv", delimiter=";")
```

# Formatando Datasets

---

Os dados apresentados não estão em um formato aceitável para análise.

Foi feito:

- remoção do código de municipio e inseri-lo em uma coluna separada

```python

try:
    # Drop da coluna municipio/codigo original
    mun_col_cito = df_cito_siscan.pop('municipio_de_saude')
    mun_col_mamo = df_mamo_siscan.pop('municipio_de_saude')

    # Separacao de codigo e muicipio
    df_cito_siscan[['codigo_mun', 'municipio']] = mun_col_cito.str.extract(r'(\d+)\s+(.+?)\s*$')
    df_mamo_siscan[['codigo_mun', 'municipio']] = mun_col_mamo.str.extract(r'(\d+)\s+(.+?)\s*$')

except KeyError:
    print('Col municipio_de_saude já removido')

finally:
    print(df_cito_siscan.head())
    print(df_mamo_siscan.head())
```

```python
df_citopatologia_cariri  = df_cito_siscan[
    (df_cito_siscan['municipio'].isin(cidades_cariri)) 
]
df_mamografia_cariri  = df_mamo_siscan[
    (df_mamo_siscan['municipio'].isin(cidades_cariri)) 
]
```

```python
df_citopatologia_cariri
```

```python
df_mamografia_cariri
```

# Reorganização dos dados

---
A partir daqui serão tratados dados inconsistentes e preparação para novos datasets

```python
df_long_cito = df_citopatologia_cariri.melt(
    id_vars=['municipio'],
    value_vars=[str(ano) for ano in range(2016, 2025)],
    var_name='ano',
    value_name='valor_indicador'
)

df_long_mamo = df_mamografia_cariri.melt(
    id_vars=['municipio'],
    value_vars=[str(ano) for ano in range(2016, 2025)],
    var_name='ano',
    value_name='valor_indicador'
)
```

```python
df_long_cito
```

```python
df_long_cito['ano_mes'] = df_long_cito['ano'] + '12'
df_long_cito['estado'] = 'CE'
df_long_cito['tipo_exame'] = 'Citopatologico'

df_long_mamo['ano_mes'] = df_long_cito['ano'] + '12'
df_long_mamo['estado'] = 'CE'
df_long_mamo['tipo_exame'] = 'Mamografia'
```

```python
df_long_cito
```

# Limpeza final

---

A partir daqui, o dataset cito_e_mamo.csv será atualizado para poder fazer um merge aos datasets atuais.
Os Datasets atuais também precisam ter certas colunas removidas antes de serem salvos

```python
colunas = ['estado', 'municipio', 'valor_indicador', 'ano_mes', 'tipo_exame']

df_cito = df_long_cito[colunas]
df_mamo = df_long_mamo[colunas]
```

```python
# Carregando cito_e_mamo.csv
df_cito_e_mamo_sus = pd.read_csv('../../Datasets/N1/cito_e_mamo_cariri.csv')

import unicodedata
# Necessario 
def normalizar(txt):
    if pd.isna(txt):
        return txt
    txt = str(txt).strip().upper()
    txt = unicodedata.normalize('NFKD', txt)
    txt = ''.join(c for c in txt if not unicodedata.combining(c))
    return txt

df_cito_e_mamo_sus['municipio'] = df_cito_e_mamo_sus['municipio'].apply(normalizar)

try:
    df_cito_e_mamo_sus = df_cito_e_mamo_sus.drop(columns='data_atualizacao')
except KeyError:
    print('Col já removida')
```

```python
df_cito_e_mamo_sus
```

# Merge de Datasets

---
Ocorretá agora o merge entre o dataset cito_e_mamo.csv, mamografia_atendbrCE.csv e cito_colo_atendbrCE.csv seguindo as regras abaixo:

- Linhas com valores anomes repetidas terão seus valores comparados, o valor com maior quantidade será substituido 
- Linhas com valores anomes não existentes em cito_e_mamo que estão presentes nos demais datasets serão adicionados


```python
# Merge de mamografia_atendbrCE.csv e cito_colo_atendbrCE.csv

df_cito_e_mamo_siscan = pd.concat(
    [df_cito, df_mamo],
    ignore_index=True
)

df_cito_e_mamo_siscan = df_cito_e_mamo_siscan[df_cito_e_mamo_siscan['valor_indicador'] != 0]

df_cito_e_mamo_siscan['ano_mes'] = df_cito_e_mamo_siscan['ano_mes'].astype(int)

df_cito_e_mamo_siscan
```

```python
# Concatecacao
df_concatenado_siscan_sus = pd.concat([
    df_cito_e_mamo_siscan, df_cito_e_mamo_sus
], ignore_index=True)


df_concatenado_siscan_sus

```

```python
# Agrupacao Final

df_final = (
    df_concatenado_siscan_sus
    .groupby(['estado', 'ano_mes', 'municipio', 'tipo_exame'], as_index=False)
    .agg({'valor_indicador': 'sum'})
)

```

```python
df_final
```

# Salvamento dos Datasets

---
Os dados agora serão registrados em CSV.


```python
df_final.to_csv('../../Datasets/N1/df_cito_e_mamo_sus_siscan.csv')
df_cito_e_mamo_siscan.to_csv('../../Datasets/N1/cito_e_mamo_siscan.csv')
df_cito.to_csv('../../Datasets/N1/cito_siscan.csv')
df_mamo.to_csv('../../Datasets/N1/mamo_siscan.csv')
```
