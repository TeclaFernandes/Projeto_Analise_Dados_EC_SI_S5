### Códigos

As variáveis abaixo são todas Qualitativas Nominais, pois principalmente, classificam cada linha por seus códigos.
```
- co_ibge: Código IBGE do município.
- co_uf: Código da Unidade da Federação.
- co_regiao_saude: Código da região de saúde.
- co_macro: Código da macrorregião.
``` 


Quanto a essa variavel, co_anomes, é uma ordinal categorica, principalmente por ser uma combinação entre Ano + mês.
```
- co_anomes: Código no formato Ano + Mês (Ex.: 202301 => Janeiro de 2023)
```

````python

codigos_nome = {
'co_ibge': 'Codigo_IBGE', 
 'co_uf': 'Codigo_unidade_federacao', 
 'co_regiao_saude': 'Codigo_Regiao_saude',
 'co_macro': 'Codigo_macrorregiao',
 'co_anomes': 'Codigo_Ano_Mes'}

bronze_df.rename(columns=codigos_nome)[['Codigo_IBGE', 'Codigo_unidade_federacao', 'Codigo_Regiao_saude', 'Codigo_macrorregiao', 'Codigo_Ano_Mes']]
````
