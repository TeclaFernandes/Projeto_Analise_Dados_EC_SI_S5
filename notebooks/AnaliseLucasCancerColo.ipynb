## **Organização dos Dados**
Os dados estão estruturados como uma série temporal e geográfica (painel). Eles representam indicadores calculados em diferentes níveis (município, região de saúde, estado, macrorregião e país).

## **Variáveis Independentes**

São as variáveis que definem o cenário ou a localização do dado. Elas não mudam em função do indicador, mas servem para agrupá-lo:

**Temporais:** `co_anomes, dt_competencia` (Mês e ano da competência).

**Geográficas/Categorias:** `no_municipio` (nome), `sg_uf` (sigla do estado), `no_regiao_brasil` (região), `no_regiao_saude, no_macro`. Essas variáveis são fundamentais para filtros e agrupamentos.

## **Variáveis Dependentes**

São os valores calculados que queremos analisar ou prever:

`vl_indicador_calculado_mun`: O valor principal para o município.

`vl_indicador_calculado_rs`,`vl_indicador_calculado_uf`, `vl_indicador_calculado_br`: O valor consolidado em níveis maiores (Região de Saúde, Estado e Brasil, respectivamente).




```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('ccmec25a64.csv')

df['dt_competencia'] = pd.to_datetime(df['dt_competencia'])

# Ordenar os dados por data
df = df.sort_values(by='dt_competencia')

# Agrupamos por data e pegamos a média do indicador nacional
time_series = df.groupby('dt_competencia')['vl_indicador_calculado_br'].mean().reset_index()

plt.figure(figsize=(12, 6))
sns.lineplot(data=time_series, x='dt_competencia', y='vl_indicador_calculado_br', marker='o', color='royalblue')

plt.title('Evolução Temporal do Indicador Previne Brasil (Nível Nacional)', fontsize=14)
plt.xlabel('Data de Competência', fontsize=12)
plt.ylabel('Valor do Indicador', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('evolucao_temporal.png')
plt.show()
```

# **Evolução Temporal:**

Mostra como o indicador nacional `(vl_indicador_calculado_br)` variou ao longo dos anos.

```python
# Pegamos apenas os dados do último mês disponível no arquivo
ultima_data = df['dt_competencia'].max()
df_recente = df[df['dt_competencia'] == ultima_data]

# Soma dos indicadores municipais por região
region_data = df_recente.groupby('no_regiao_brasil')['vl_indicador_calculado_mun'].sum().sort_values(ascending=False).reset_index()

plt.figure(figsize=(10, 6))
sns.barplot(data=region_data, x='vl_indicador_calculado_mun', y='no_regiao_brasil', palette='viridis')

plt.title(f'Soma do Indicador por Região ({ultima_data.strftime("%m/%Y")})', fontsize=14)
plt.xlabel('Soma dos Valores Municipais', fontsize=12)
plt.ylabel('Região', fontsize=12)
plt.tight_layout()
plt.savefig('soma_por_regiao.png')
plt.show()
```
